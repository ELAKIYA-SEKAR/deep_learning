```python
def log_loss(y_true,y_pred):
  epsilon=1e-15
  y_pred_new=[max(i,epsilon) for i in y_pred]
  y_pred_new=[min(i,1-epsilon) for i in y_pred_new]
  y_pred_new=np.array(y_pred_new)
  return -np.mean(y_true*np.log(y_pred_new)+(1-y_true)*np.log(1-y_pred_new))
```

```python
def sigmoid_numpy(x): #vector sigmoid
  return 1/(1+np.exp(-x))

sigmoid_numpy(np.array([34,5,6]))
```

```python
def gradient(age,affordability,y_true,epochs):
  w1=w2=1
  bias=0
  learning_rate=0.5
  n=len(age)

  for i in range(epochs):
    weighted_sum= w1*age+w2*affordability + bias
    y_pred = sigmoid_numpy(weighted_sum)

    loss = log_loss(y_true,y_pred)

    w1d=(1/n)*np.dot(np.transpose(age),(y_pred-y_true))
    w2d=(1/n)*np.dot(np.transpose(affordability),(y_pred-y_true))

    bias=np.mean(y_pred-y_true)

    w1=w1-learning_rate*w1d
    w2=w2-learning_rate*w2d
    bias=bias-learning_rate*bias

    print("epoch",epochs,"w1:",w1,"w2:",w2,"bias:",bias,"loss:",loss)
  return w1,w2,bias
```
