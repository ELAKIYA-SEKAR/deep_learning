```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/breast-cancer.csv")
```

```python
df.sample(10)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoding=LabelEncoder()
df['diagnosis']=label_encoding.fit_transform(df['diagnosis'])
```

```python
df
```

```python
df.drop('id',axis=1)
```

```python
x=df.drop('diagnosis',axis=1)
y=df.diagnosis
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
parameters={'criterion':('gini', 'entropy', 'log_loss'),'max_features':('sqrt', 'log2', None),'class_weight':('balanced', 'balanced_subsample')}
grid_search=GridSearchCV(rfc,parameters)
grid_search.fit(x_train,y_train)
data_frame=pd.DataFrame(grid_search.cv_results_)
print(data_frame)
print("Best Parameters:",grid_search.best_params_)
print("Best score:",grid_search.best_score_)
```

```python
from sklearn.metrics import accuracy_score
y_pred=grid_search.predict(x_test)
print("Accuracy Score of Random FOrest",accuracy_score(y_test,y_pred))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gbc=GradientBoostingClassifier()
params={'loss':('log_loss', 'exponential'),'criterion':('friedman_mse', 'squared_error'),'max_features':('sqrt','log2',None)}
grid_gb=GridSearchCV(gbc,params)
grid_gb.fit(x_train,y_train)
df1=pd.DataFrame(grid_gb.cv_results_)
print(df1)
print("Best parameters:",grid_gb.best_params_)
print("Best score:",grid_gb.best_score_)
```

```python
y_pred1=grid_gb.predict(x_test)
print("Accuracy Score of Gradient Boosting",accuracy_score(y_test,y_pred1))
```

```python
from sklearn.ensemble import AdaBoostClassifier
ada=AdaBoostClassifier()
params1={'algorithm':('SAMME', 'SAMME.R'),'n_estimators':[1,5,10,20]}
grid_ada=GridSearchCV(ada,params1)
grid_ada.fit(x_train,y_train)
df2=pd.DataFrame(grid_ada.cv_results_)
print(df2)
print("Best parameters:",grid_ada.best_params_)
print("Best score:",grid_ada.best_score_)
```

```python
y_pred2=grid_ada.predict(x_test)
print("Accuracy Score of Ada Boost",accuracy_score(y_test,y_pred2))
```

```python
from xgboost import XGBClassifier
xgb=XGBClassifier()
xgb.fit(x_train,y_train)
y_pred3=xgb.predict(x_test)
print("Accuracy Score of XG Boost",accuracy_score(y_test,y_pred3))
```
